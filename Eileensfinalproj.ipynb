```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
import csv
```

```python
UninsuredData = pd.read_csv('uninsuredData.csv')
print(UninsuredData.head())
```

```python
def plotUninsuredDataBar(data):

    data['Washington'] = data['Washington'].str.replace('%', '').astype(float)
    data['Texas'] = data['Texas'].str.replace('%', '').astype(float)


    bar_width = 0.35
    index = data['Year']


    plt.bar(index - bar_width/2, data['Washington'], bar_width, label='Washington', color='green')
    plt.bar(index + bar_width/2, data['Texas'], bar_width, label='Texas', color='orange')

    plt.xlabel('Year')
    plt.ylabel('Percentage of Uninsured People')
    plt.title('Uninsured Percentage by Year (Washington vs Texas)')
    plt.legend()
    plt.grid(True)
    plt.show()


UninsuredData = pd.read_csv('uninsuredData.csv')
plotUninsuredDataBar(UninsuredData)
```
