# Uninsured Patients Data
I chose this graph to visually represent the percentage of uninsured patients because it highlights a critical aspect of healthcare access disparities. The percentage of uninsured individuals serves as a clear indicator of how accessible healthcare is to different populations. By focusing on uninsured rates, the graph allows for a straightforward comparison of healthcare coverage between groups, regions, or time periods. This metric is particularly important because it directly correlates with health outcomes and equity, making it an effective choice for examining public health disparities.

```python
UninsuredData = pd.read_csv('uninsuredData.csv')
print(UninsuredData.head())
```

```python
def plotUninsuredDataBar(data):

    data['Washington'] = data['Washington'].str.replace('%', '').astype(float)
    data['Texas'] = data['Texas'].str.replace('%', '').astype(float)


    bar_width = 0.35
    index = data['Year']


    plt.bar(index - bar_width/2, data['Washington'], bar_width, label='Washington', color='green')
    plt.bar(index + bar_width/2, data['Texas'], bar_width, label='Texas', color='orange')

    plt.xlabel('Year')
    plt.ylabel('Percentage of Uninsured People')
    plt.title('Uninsured Percentage by Year (Washington vs Texas)')
    plt.legend()
    plt.grid(True)
    plt.show()


UninsuredData = pd.read_csv('uninsuredData.csv')
plotUninsuredDataBar(UninsuredData)
```

# Medicare Enrollees Receiving FLu Vaccinations
I chose this graph because it effectively visualizes flu vaccination trends among Medicare enrollees in Washington and Texas over the years, which is an important public health indicator. By plotting the data for these two states together, it allows for a clear comparison of vaccination rates and highlights regional differences in public health efforts. The use of distinct colors, green for Washington and orange for Texas, makes it easy to distinguish between the states, while the markers help emphasize each year's data points in a straightforward way.

The decision to convert the percentage strings into floats ensures the data is in a numerical format suitable for plotting, which maintains both accuracy and clarity in the visual representation. This approach also avoids errors that could occur from treating the percentages as text. Including a grid and a legend adds to the readability of the graph, making it easier for viewers to interpret the trends and differences.

I chose the percentage of enrollees receiving flu vaccinations as the focus because it reflects healthcare accessibility and the prioritization of public health initiatives aimed at preventing seasonal flu outbreaks. Flu vaccination rates are a valuable metric for assessing the effectiveness of health campaigns and identifying gaps in healthcare delivery. This makes the graph a useful tool for analyzing and discussing healthcare disparities between these states.

```python
fluVaccinations = pd.read_csv('fluVaccinations.csv')
print(fluVaccinations.head())
```

```python

fluVaccinations['Washington'] = fluVaccinations['Washington'].astype(str).str.rstrip('%').astype(float)
fluVaccinations['Texas'] = fluVaccinations['Texas'].astype(str).str.rstrip('%').astype(float)
plt.plot(fluVaccinations['Year'], fluVaccinations['Washington'], label='Washington', marker='o',color='green')
plt.plot(fluVaccinations['Year'], fluVaccinations['Texas'], label='Texas', marker='o',color='orange')


plt.title('Medicare Enrollees Receiving Flu Vaccinations')
plt.xlabel('Year')
plt.ylabel('Enrollees Receiving Flu Vaccination Rate (%)')
plt.legend()
plt.grid(True)
plt.show()

```

# Income Inequality Data

This histogram illustrates the income inequality ratios in Texas and Wasington in the 80th Percentile (High-income threshold for the wealthier population) and in the 20th Percentile (Low-income threshold for the poorer population) of the population. The x-axis compares the income ranges between the 20th and 80th percentiles. The y-axis represents the distribution density and the frequency of counties from both states that fall within certain income ranges. Different colors were used to distinguish between different percentiles.
The first thing that is noticeable after analyzing the histogram is the y-axis. Since Washington has fewer counties than Texas the values on the y-axis for Washington are less than Texas. There are greater gaps between the 20th and 80th percentile in the Texas graph which means Texas has greater income inequality than Washington. Taking a look at the distribution of data in the x-axis of both graphs it looks like the population in Washington has more people making more money than in Texas. Overall Texas has a higher gap between incomes and they make less income than the state of Washington. 
Income inequality has a big effect on healthcare. Access to quality healthcare can be easier and more accessible to the wealthy while the lower income population struggles to get basic healthcare. Wealthier areas tend to have more hospitals making it easier for patients to be seen. On the other hand, lower income populations have fewer hospitals making it difficult to get undivided attention.  


```python
texas_file = pd.read_csv('texas_file - texas-2024-income-inequality-place-sort.csv.csv')
washington_file = pd.read_csv('washington_file - washington-2024-income-inequality-place-sort.csv.csv')
print(texas_file)
print(washington_file)
```

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
import csv

def get_data_df(df):
    percentile_80 = df.iloc[:, 1].str.replace(',', '').astype(float).tolist()
    percentile_20 = df.iloc[:, 2].str.replace(',', '').astype(float).tolist()
    return percentile_80, percentile_20
    
def plot_histograms(texas_df, washington_df):
    texas_80, texas_20 = get_data_df(texas_df)
    washington_80, washington_20 = get_data_df(washington_df)
    plt.figure(figsize=(14, 7))

    plt.subplot(1, 2, 1)
    plt.hist(texas_20, bins=20, alpha=0.5, label='Texas 20th Percentile', color='purple')
    plt.hist(texas_80, bins=20, alpha=0.5, label='Texas 80th Percentile', color='blue')
    plt.title('Income Distrubution Ratio')
    plt.xlabel('Income (in dollars)'
    plt.ylabel('Frequency (')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.hist(washington_80, bins=20, alpha=0.5, label='Washington 80th Percentile', color='green')
    plt.hist(washington_20, bins=20, alpha=0.5, label='Washington 20th Percentile', color='red')
    plt.title('Income Distrubution Ratio')
    plt.xlabel('Income')
    plt.ylabel('Frequency')
    plt.legend()

    plt.tight_layout()
    plt.show()


plot_histograms(texas_file, washington_file)
```

# Children in Poverty 2024

The boxplot below shows Texas and Washingtons distribution of child poverty rates within their respective counties. The x-axis represents the states while the y-axis represents the poverty rates across the counties. Boxplots are useful when comparing two variables. They are also visibly easy to read. The different elements in a boxplot are the median, range, and interquartile range(IQR). 

After studying the boxplot there are a few things that stand out. One of them being outliers, the Texas boxplot seems to have outliers while the Washington one does not. This could be to the fact that Texas has more counties and a greater gap in income. The median rate of Texas poverty is higher than that of Washington. Both of these observations prove that Washington is better off when it comes to income within its population.

Children's poverty rates affect healthcare. Poverty is often generational and difficult to get out of. When children grow up in poverty it is difficult for them to get preventative care and hard to access services to help with early detection of serious illnesses. Preventative care and early detection can then cause early mortality. 



```python
texas_df = pd.read_csv('texas_child_poverty - texas-2024-children-in-poverty-place-sort.csv (2).csv')
washington_df = pd.read_csv('washington_child_poverty - washington-2024-children-in-poverty-place-sort.csv (3).csv')
print(texas_df.head())
print(washington_df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

texas_df.columns = texas_df.columns.str.strip().str.lower()
washington_df.columns = washington_df.columns.str.strip().str.lower()

texas_df['state'] = 'Texas'
washington_df['state'] = 'Washington'

texas_df['county value'] = pd.to_numeric(texas_df['county value'].str.rstrip('%'), errors='coerce') 
washington_df['county value'] = pd.to_numeric(washington_df['county value'].str.rstrip('%'), errors='coerce')

combined_data = pd.concat([texas_df[['state', 'county value']], washington_df[['state', 'county value']]])

plt.figure(figsize=(8, 6))
combined_data.boxplot(by= 'state', column='county value', grid=False, patch_artist=True,
                      boxprops=dict(facecolor='green', color='purple'))

plt.title('Children Poverty Rates by State', fontsize=16)
plt.suptitle('')
plt.xlabel('State', fontsize=18)
plt.ylabel('Poverty Rate (%)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
